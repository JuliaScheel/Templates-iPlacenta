## Data Types

### Every value in a program has a specific type.
* Integer (int): represents positive or negative whole numbers like 3 or -512.
* Floating point number (float): represents real numbers like 3.14159 or -2.5.
* Character string (usually called “string”, str): text.
    * Written in either single quotes or double quotes (as long as they match).
    * The quote marks aren’t printed when the string is displayed.
    
### Use the built-in function type to find the type of a value. 

```python
print(type(52))
```

```python
fitness = "average"
print(type(fitness))
```

### Types control what operations (or methods can be performed on a given value.

* A value’s type determines what the program can do to it

```python
print(5-3)
```

```python
print("hello"-"h")
```

### You can use the <span style="color:red">+</span> and <span style="color:red">*</span> operators on strings

```python
full_name = "Julia" + " " + "Scheel"
print(full_name)
```

### Strings have a length - numbers do not

```python
print(len(full_name))
```

```python
print(len(52))
```

### You must convert numbers to strings or vice versa when operating on them

```python
print(1 + "2")
```

### Convert a type into another type using <span style="color:red">int</span> or <span style="color:red">str</span> commands

```python
print(1 + int("2"))
```

```python
print(str(1)+"2")
```

### Variables only change value when something is assigned to them

```python
first = 1
second = 5 * first
first = 2
print("first is", first, "and second is", second)
```

##  Reading Tabular Data into DataFrames (df)

* import pandas library
* use pandas to load a csv data set
* get basic information about data set

```python
import pandas as pd

df = pd.read_csv("gapminder_gdp_oceania.csv")
df.head()
```

### Use <span style="color:red">index_col </span>to specify that a column's values should be used as row headings

```python
df = pd.read_csv("gapminder_gdp_oceania.csv", index_col = "country")
df.head()
```

### Use <span style="color:red">DataFrame.info</span> to find out more about a dataframe

```python
df.info()
```

### Use  <span style="color:red">DataFrame.T </span> to transpose a dataframe

```python
print(df.T)
```

### Use  <span style="color:red">DataFrame.describe</span>  to get summary statistics about data

```python
df.describe()
```

### Use <span style="color:red"> DataFrame.iloc[..., ...]</span>  to select values by their position

```python
data = pd.read_csv('gapminder_gdp_europe.csv', index_col='country')
print(data.iloc[0, 0])
```

```python
data.head()
```

### Use <span style="color:red"> DataFrame.loc[..., ...] </span> to select values by their label

```python
print(data.loc["Albania", "gdpPercap_1952"])
```

### Use <span style="color:red"> : </span> on its own to select all columns or all rows

```python
print(data.loc["Albania", :])
```

```python
print(data.loc[:, "gdpPercap_1952"])
```

### Select multiple columns or rows using  <span style="color:red"> DataFrame.loc</span> and a  <span style="color:red"> named slice</span>.

```python
print(data.loc['Italy':'Poland','gdpPercap_1962':'gdpPercap_1972'])
```

```python
print(data.loc['Italy':'Poland', 'gdpPercap_1962':'gdpPercap_1972'].max())
```

## Intro to Plotting in Python

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
time = [0, 1, 2, 3]
position = [0, 100, 200, 300]

plt.plot(time, position)
plt.xlabel('Time (hr)')
plt.ylabel('Position (km)')
```

```python
df.head()
```

We can also plot a pandas dataframe
In this case we have to convert the column headings from a <span style="color:red">string</span> to <span style="color:red">integer</span> data type. We want to extract the last 4 characters of each column name.
We can use the <span style="color:red">strip()</span> (only works on strings). 

```python
years = df.columns.str.strip("gdpPercap_")
df.columns = years.astype(int)
df.head()
```

```python
df.loc["Australia"].plot()
```

```python
df.T.plot()
plt.ylabel("GDP per capital")
```

Many styles of plot are available.

```python
plt.style.use('ggplot')
df.T.plot(kind='bar')
plt.ylabel('GDP per capital')
```

### Plot dashed line

```python
#years = data.columns
gdp_australia = df.loc['Australia']

plt.plot(years, gdp_australia, 'g--')
```
